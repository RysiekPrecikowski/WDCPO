```python
def filter(img, ddepth=-1, kernel=np.ones((3, 3))/9, plot=True):
    res = cv2.filter2D(img, ddepth, kernel)

    if plot:
        print("original")
        plt.imshow(img, cmap ="gray")
        plt.show()
        print("filtered")
        plt.imshow(res, cmap="gray")
        plt.show()
    else:
        return res

filter(plansza)
```

```python
def test_filtration(img, kernel=lambda i: np.ones((i, i))/(i*i)):
    print("original")
    plt.imshow(img, cmap="gray")
    plt.show()
    for i in [3,5,8,15,35]:
        filtered = filter(img, kernel=kernel(i), plot=False)
        print(f"{i}x{i}")
        plt.imshow(filtered, cmap="gray")
        plt.show()

test_filtration(plansza)
```

```python
def fgaussian(size, sigma):
     m = n = size
     h, k = m//2, n//2
     x, y = np.mgrid[-h:h+1, -k:k+1]
     g = np.exp(-(x**2 + y**2)/(2*sigma**2))
     return g /g.sum() 
    
    
def mesh(fun, size):
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    

    X = np.arange(-size//2, size//2, 1)
    Y = np.arange(-size//2, size//2, 1)
    X, Y = np.meshgrid(X, Y)
    Z = fun
    
    ax.plot_surface(X, Y, Z)
    
    plt.show()

print(fgaussian(5, 0.5))

filter(lena, kernel=fgaussian(35, 0.5))
test_filtration(lena, kernel=lambda i: fgaussian(i, 0.5))
```
